```python
# Import de bibliotecas e dados
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import sys
sys.path
sys.executable
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
# Load dataset
df = pd.read_csv('datasets/cc_approvals.data',header=None) 
df.head()
```

```python
#df.describe()
df.info()
```

```python
df.tail(20)  
```

```python
df = df.replace('?',np.nan)
```

```python
df.tail(20)
```

```python
df.fillna(df.mean(), inplace=True)
df.isnull().sum()
```

```python
for col in df.columns:
    # Checar se a coluna é do tipo object 
    if df[col].dtypes == 'object':
        # Imputar com os valores mais frequentes
        df = df.fillna(df[col].value_counts().index[0])
df.isnull().sum()
```

```python
le = LabelEncoder()
```

```python
for col in df.columns:
    # Compare se o the dtype é object
    if df[col].dtypes=='object':
    # Use LabelEncoder para fazer a transformação numérica
        df[col]=le.fit_transform(df[col])
df.describe()        
```

```python
# Drop the features 11 and 13 and convert the DataFrame to a NumPy array
df = df.drop([df.columns[11], df.columns[13]], axis=1)
df = df.values
X = df[:,0:13] 
y = df[:,13]
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.33, random_state=42)
X_treino
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
rescaledX_treino = scaler.fit_transform(X_treino)
rescaledX_teste = scaler.fit_transform(X_teste)
```

```python
# Import LogisticRegression
# Instantiate a LogisticRegression classifier with default parameter values
logreg = LogisticRegression()
# Fit logreg to the train set
logreg.fit(rescaledX_treino,y_treino)
```

```python
#Matriz de Confusão

y_pred = logreg.predict(rescaledX_teste) # Use logreg to predict instances from the test set and store it
print(" Precisão do Classificador é: ", logreg.score(rescaledX_teste,y_teste))
print(confusion_matrix(y_teste,y_pred))  #  matriz de confusão do modelo logreg
```

```python
tol = [0.01,0.001,0.0001]
max_iter = [100,150,200]
param_grid = dict(tol=tol,max_iter=max_iter)
print(param_grid)
```

```python
grid_model = GridSearchCV(estimator=logreg, param_grid=param_grid, cv=5)
rescaledX = scaler.fit_transform(X)
grid_model_result = grid_model.fit(rescaledX, y)
best_score, best_params = grid_model_result.best_score_,grid_model_result.best_params_
print("Best: %f using %s" % (best_score, best_params))
```
